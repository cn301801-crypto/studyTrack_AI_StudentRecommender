```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("combined_df_with_clusters.csv")
df.head()
```

```python
#Check cluster distribution
print("Unique clusters:", df["Cluster"].unique())
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Cluster", data=df)
plt.title("Cluster Distribution")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()
```

```python
# Define recommendations and tools/resources for each cluster
def get_recommendation(cluster):
    if cluster == 0:
        return "Focus on improving study habits and reduce social distractions."
    elif cluster == 1:
        return "Maintain consistent effort; try time management techniques."
    elif cluster == 2:
        return "Great academic potential; explore leadership or mentor roles."
    else:
        return "General improvement needed; seek academic counseling."
def get_tool_or_resource(cluster):
    if cluster == 0:
        return "Use Pomodoro timer apps, basic revision planners."
    elif cluster == 1:
        return "Try Google Calendar, Trello for tracking assignments."
    elif cluster == 2:
        return "Explore Coursera or advanced subject-specific projects."
    else:
        return "Join extra help sessions and use YouTube learning resources."

# Apply both functions to your dataset
df["Recommendation"] = df["Cluster"].apply(get_recommendation)
df["Suggested_Tool_or_Resource"] = df["Cluster"].apply(get_tool_or_resource)
df.head()
```

```python
import os
os.makedirs("visualizations", exist_ok=True)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
sns.countplot(y="Recommendation", data=df, palette="viridis")
plt.title("Number of Students per Recommendation")
plt.xlabel("Count")
plt.ylabel("Recommendation")
plt.tight_layout()
plt.savefig("visualizations/recommendation_countplot.png")
plt.show()
print("Visualization saved as 'visualizations/recommendation_countplot.png'")
```

```python
df.to_csv("student_recommendations.csv", index=False)
print("File saved successfully as 'student_recommendations.csv'")
```
